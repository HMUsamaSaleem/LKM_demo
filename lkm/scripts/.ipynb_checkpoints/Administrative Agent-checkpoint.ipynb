###### IMp

```python
df = pd.read_csv("C:/Users/PC/Desktop/pixeledge/lkm/dataset/inputfile/23oct.csv")
driver = webdriver.Firefox()
first_10_lines_list = []

for index, row in df.iterrows():
    url = row["SEC.gov URL"]
    driver.get(url)
    sleep(5)
    page_text = driver.find_element(By.TAG_NAME, 'body').text
    first_10_lines = page_text.splitlines()[:15]
    first_10_lines_list.append(first_10_lines)

driver.quit()
```

```python
# Initialize the QA model
model_name = "deepset/roberta-base-squad2"
nlp = pipeline('question-answering', model=model_name, tokenizer=model_name)
for lines in first_10_lines_list:
    # Convert the lines to a single string (joining each line with a space)
    context = " ".join(lines)

    # Define your question
    question = "what is the name of admministrative agent?"

    # Use the QA model to extract the answer
    QA_input = {'question': question, 'context': context}
    result = nlp(QA_input)
    if result['score'] < 0.5:
        question = "Who is the administrative agent in this text?"
        QA_input = {'question': question, 'context': context}
        result = nlp(QA_input)
    if result['score'] < 0.5:
        question = "Who serves as the administrative agent?"
        QA_input = {'question': question, 'context': context}
        result = nlp(QA_input)
    if result['score'] < 0.5:
        question = "Who holds the position of administrative agent?"
        QA_input = {'question': question, 'context': context}
        result = nlp(QA_input)
    if result['score'] < 0.5:
        lines = str(lines)
        if "Administrative Agent" in lines:
            text_1 = lines.split('as Administrative Agent')
            Agent_name = re.findall(r'\b[A-Z][A-Z\s,]+\b', str(text_1[0]))
            print(Agent_name[-1])
        elif "administrative agent" in lines:
            text_1 = lines.split('as administrative agent')
            Agent_name = re.findall(r'\b[A-Z][A-Z\s,]+\b', str(text_1[0]))
            print(Agent_name[-1])
        elif "as agent" in lines:
            text_1 = lines.split('as agent')
            Agent_name = re.findall(r'\b[A-Z][A-Z\s,]+\b', str(text_1[0]))
            print(Agent_name[-1])
        else:
            print(None)
            print("##################")
        
    # Print the answer and other relevant information
    print(f"Question: {question}")
    print(f"Answer: {result['answer']}")
    print(f"Confidence: {result['score']}")  # Confidence score
    print()
```
